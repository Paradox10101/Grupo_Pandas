```python
from ultralytics import YOLO
```

```python
model2 = YOLO("yolov8n.yaml")
# Anteriormente se realizo una ejecución de entrenamiento de 50 epocas y se guardaron los pesos en el archivo primerbest.pt

# Cargar los pesos guardados en el archivo primerbest.pt
model2.load("/content/primerbest.pt")
```

```python
model2.train(data="config.yaml", epochs=25) # Entrenar por 25 epocas adicionales
# Guardar los pesos después del segundo entrenamiento
model2.save('second_model.pt')
```

```python
metrics2 = model2.val() # Se obtienen las métricas de validación
```

```python
results = model2("papel.jpg")  # predice la imagen papel.jpg
path = model2.export(format="onnx")  # exporta el modelo a formato ONNX
```

```python
results = model2("glass21.jpg")  # predice la imagen papel.jpg

import matplotlib.pyplot as plt
import cv2
# Asegúrate de que 'results' no esté vacío y es del tipo esperado
if results and isinstance(results, list):
    for result in results:
        # Renderizar la imagen con las predicciones
        img_with_boxes = result.plot()

        # Convertir la imagen de BGR a RGB (si es necesario)
        img_with_boxes_rgb = cv2.cvtColor(img_with_boxes, cv2.COLOR_BGR2RGB)

        # Mostrar la imagen
        plt.imshow(img_with_boxes_rgb)
        plt.axis('off')  # Ocultar los ejes
        plt.show()
else:
    print("No se encontraron resultados.")
```
